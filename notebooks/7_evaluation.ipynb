## 7. Evaluation

This notebook is just a written summary of the whole project. We already produced all the metrics, plots and per class breakdowns in notebooks 3-6. I want to wrap up what we did, what worked, where the limits are and what I would try next.

## 7.1 Results

The CNN clearly outperforms the random guessing benchmark and the simple ANN we compared in notebook 3. So there are learnable visual patterns that the model can pick up to tell the rarity classes apart.

The grid search in notebook 6 found a small improvement over the notebook 4 baseline by using a slightly stronger L2, but the difference is inside the noise band of multi-run averaging from notebook 4. So the techniques themselves are fine, the values were not the bottleneck and no config really breaks past the same accuracy ceiling.

For a production ready system the accuracy is still not good enough, but for this learning project I think the result is solid.

## 7.2 Interpretation

Looking at the confusion matrix and the per class metrics, the errors are not random:

- **Common, Uncommon and Rare** get mixed up heavily. They only differ in a tiny symbol on the bottom of the card (circle, diamond, star), which is barely visible at our 168x120 resolution.
- **Ultra Rare** is the strongest class. The holographic effects and the different layout (full card pokemon illustration) make it easy to spot.
- **Secret Rare** is also strong. Rainbow, gold and full art designs are visually unique enough that the CNN catches them reliably.

So the model has clearly learned the "looks holographic / has full art" axis, but it cant resolve the small rarity symbols that separate the lower three classes.

## 7.3 Limitations

- **Resolution.** 168x120 was chosen because of hardware constraints, but at that size the rarity symbols are essentially washed out.
- **Dataset size.** I used 2500 images (500 per class) instead of the full ~18.000, again because of hardware and time. This is just not enough for a CNN to learn subtle features from scratch.
- **Holographic effects.** Static low resolution images dont really show the shiny / holographic effect that is so important on a real card.
- **Overfitting.** Training accuracy is much higher than validation accuracy. I tried to close this gap with regularization (notebook 4), architecture changes like augmentation and GAP (notebook 5) and a small grid search over dropout and L2 values (notebook 6), but none of them really helped. So the bottleneck is not the model but the data.

## 7.4 Possible Improvements

Things I would try next if I had better hardware and more time:

- **Higher resolution images** so the small rarity symbols on the bottom are actually visible
- **Cropping the bottom of the card** so the model focuses on the area where the rarity symbol sits, instead of looking at the whole card
- **More data**, either by using more of the 18.210 cards from the original dataset or by collecting additional images
- **Combining image features with card metadata** (name, type, set), since this information is already in the dataset and could push accuracy further

I already tried some of these but my hardware (8GB M1) crashed on the higher resolution runs, so this stays out of scope for the current setup.

## 7.5 Conclusion

The CNN learns clearly meaningful patterns and clearly beats the benchmarks, but the gap between training and validation accuracy and the per class confusion show a classic small dataset problem. The model has enough capacity to memorize but not enough data and resolution to generalize on the visually similar Common, Uncommon and Rare classes. Architecture and regularization tuning didnt fix this, so the next real step would be to attack the data side (resolution, cropping, more data, transfer learning) rather than the model side.
